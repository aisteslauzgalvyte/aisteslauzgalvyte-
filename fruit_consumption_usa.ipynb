```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'ID': [25, 61, 85, 109, 133, 157, 181, 205, 229, 253],
    'Age': [18, 18, 18, 18, 18, 18, 18, 18, 18, 18],
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male', 'Male'],
    'Country': ['USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA', 'USA'],
    'Favorite Fruit': ['Strawberry', 'Apple', 'Strawberry', 'Strawberry', 'Strawberry', 'Strawberry',
                        'Strawberry', 'Strawberry', 'Strawberry', 'Strawberry'],
    'Frequency of Consumption': ['Daily', '4-5 times a week', 'Once a week', 'Daily', 'Once a week', 'Once a week',
                                 'Once a week', 'Once a week', 'Once a week', 'Once a week'],
    'Preferred Form of Fruit': ['Fresh', 'Smoothie', 'Juice', 'Fresh', 'Salad', 'Juice', 'Smoothie', 'Juice', 'Smoothie', 'Fresh']
}


df = pd.DataFrame(data)

frequency_map = {
    'Daily': 7,
    '2-3 times a week': 2.5,
    'Once a week': 1,
    '4-5 times a week': 4.5,
    'NaN': None
}

df['Consumption per Week'] = df['Frequency of Consumption'].map(frequency_map)

age_18_top_10 = df[df['Age'] == 18].head(10)

favorite_fruit_count = age_18_top_10['Favorite Fruit'].value_counts()

plt.figure(figsize=(8, 5))
favorite_fruit_count.plot(kind='bar', color='skyblue')

plt.title('Most Popular Fruits Among First 10 People Aged 18')
plt.xlabel('Fruit')
plt.ylabel('Number of People')

plt.show()
```
